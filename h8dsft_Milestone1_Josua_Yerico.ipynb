*Menghilangkan Mising Value pada data Instagram*

```python
dataig_new = data_ig.drop(columns = 'category_2').dropna().reset_index(drop=True)
dataig_new
```

Mengubah type data object menjadi int/float pada kolom followers, authentic engagement, dan engagement

```python
def value_to_float(x):
    if type(x) == float or type(x) == int:
        return x
    if 'B' in x:
        if len(x) > 1:
            return float(x.replace('B', '')) * 1000000000       
        return 1000000000
    if 'M' in x:
        if len(x) > 1:
            return float(x.replace('M', '')) * 1000000
        return 1000000.0
    if 'K' in x:
        return float(x.replace('K', '')) * 1000
    return 1000

dataig_new['Followers'] = dataig_new['Followers'].apply(value_to_float)
dataig_new['Authentic engagement\r\n'] = dataig_new['Authentic engagement\r\n'].apply(value_to_float)
dataig_new['Engagement avg\r\n'] = dataig_new['Engagement avg\r\n'].apply(value_to_float)
dataig_new
```

- **Uji Hipotesis untuk Menentukan Hubungan antara Jumlah Followers dan Jumlah viewers di tiktok**

sebelum melakukan uji hipotesa pada bagian diperlukan untuk membuang outlier. hal ini karena pada analisis data sebelumnya terdapat data yang anomali (terlihat pada grafik antara jumlah likes/komen terhadap jumlah views). pada grafik tersebut terdapat jumlah viewers 4 juta tapi like dan komennya sangat sedikit.

    1. menentukan nilai skew

Pada bagian ini, data tersebut ingin dilihat apakah distribusinya normal atau tidak?

```python
datatt_new['Views avg.'].skew()
```

nilai skew yang lebih dari satu menunjukkan bahwa data tersebut memiliki distribusi yang ekstrim tidak normal. oleh karena itu diperlukan untuk membersihkan outliernya.

    2. Menentukan nilai IQR

```python
Q1 = datatt_new['Views avg.'].quantile(.25)
Q3 = datatt_new['Views avg.'].quantile(.75)
```

```python
IQR = Q3 - Q1
upper = Q3 + (3*IQR)
lower = Q1 - (3*IQR)

print('Nilai IQR:', IQR)
print('Batas Atas:', upper)
print('Batas Bawah:', lower)
```

    3. Menentukan Outlier

```python
outlier = datatt_new[(datatt_new['Views avg.'] >= upper) |
             (datatt_new['Views avg.'] <= lower)].reset_index()
outlier
```

    4. membersihkan data dari outlier

```python
datatt = datatt_new[(datatt_new['Views avg.'] <= upper) |
             (datatt_new['Views avg.'] >= lower)].reset_index()
datatt
```

Berdasarkan data tersebut didapat bahwa data tiktok bebas outlier. sehingga dapat dilakukan uji hipotesis dengan analisis yang maksimal

    5. Menentukan Confindence Interval (CI) 95%

```python
ratarata = datatt['Views avg.'].mean()
std = datatt['Views avg.'].std()

datatt_ci = np.random.normal(ratarata, std, 10000)

ci = stats.norm.interval(0.95, ratarata, std)
ci
```

    6. Analisis Uji Hipotesis

kasus: Apakah Ada hubungan yang signifikan antara jumlah followers dengan jumlah viewers pada data tiktok influencer?

    H0 : μ jumlah followers  =  μ jumlah viewers
    H1 : μ jumlah followers !=  μ jumlah viewers

```python
avg_followers = datatt['Subscribers count'].mean()
avg_viewers = datatt['Views avg.'].mean()

print('Average followers tiktok:', avg_followers)
print('Average viewers tiktok:', avg_viewers)
```

```python
t_stat, p_val = stats.ttest_ind(datatt['Subscribers count'], datatt['Views avg.'])
print('P-value:',p_val) 
print('t-statistics:',t_stat)
```

Dari hasil hipotesis tersebut didapat bahwa banyaknya jumlah viewers pada suatu konten tidak berpengaruh pada jumlah followersnya. meskipun pada grafik diatas terlihat adanya kesinabungan antara jumlah followers dan jumlah viewers.

- **Menentukan Hubungan antara Jenis Konten dengan Jumlah Followers dengan Mengunakan Uji Hipotesa Anova**

Pada uji hipotesa ini hanya diambil berapa sampel dari kolom kategori, yaitu Music, Photograph, dan cinema/actor dan actress. hal ini dikarenakan beberapa faktor sebagai berikut:
1. tiga kategori tersebut memiliki persen engagement rate tertinggi
2. tiga konten tersebut memiliki jumlah followers yang tinggi
3. dua diantara tiga konten tersebut memiliki nilai kemunculan yang sedikit sehingga bisa menjadi suatu ciri khas dari konten kreator tersebut

    H0 : tidak ada perbedaan yang signifikan antara tiga jenis konten tersebut terhadap jumlah followers
    H1 : terdapat perbedaan yang signifikan antara tiga jenis tersebut terhadap jumlah followers

```python
Music_konten = dataig_new[dataig_new['category_1'] == 'Music']['Followers']
Photography_konten = dataig_new[dataig_new['category_1']== 'Photography']['Followers']
Cinema_konten = dataig_new[dataig_new['category_1'] == 'Cinema & Actors/actresses']['Followers']

print("Average Followers of Music Content:",Music_konten.mean())
print("Average Followers of Photograph Content:",Photography_konten.mean())
print("Average Followers of Cinema Content:",Cinema_konten.mean())
```

```python
f_stat,p_value = stats.f_oneway(Music_konten, Photography_konten, Cinema_konten)
print('P-value:',p_value)
```

Berdasarkan data tersebut didapat nilai p-value kurang dari 0.05, artinya hipotesa nol pada uji ditolak. hal ini berarti bahwa tiga jenis konten tersebut terdapat perbedaan yang cukup signifikan terhadap jumlah followers. hal ini berarti adanya kontribusi jenis konten terhadap pertambahan jumlah followers. konten music, photograph, dan cinema merupakan konten yang berpotensial untuk menghasilkan jumlah followers yang banyak atau pertumbuhan followers yang cepat. lalu bagaimana dengan engagement rate dari tiap konten tersebut?


- **Menentukan hubungan jenis konten dengan engagement rate dengan mengunakan uji ANOVA**

Pada analisis ingin melihat hubungan antara tiga jenis konten dengan follower dan engagement tertinggi dengan engagement ratenya

    H0 : tidak ada perbedaan yang signifikan dari ketiga jenis konten tersebut terhadap engagement ratenya
    H1 :  terdapat perbedaan yang signifikan dari ketiga jenis konten tersebut terhadap engagement ratenya

```python
Music_engagement = dataig_new[dataig_new['category_1'] == 'Music']['Persen Engagement Rate (%)']
Photography_engagement = dataig_new[dataig_new['category_1']== 'Photography']['Persen Engagement Rate (%)']
Cinema_engagement = dataig_new[dataig_new['category_1'] == 'Cinema & Actors/actresses']['Persen Engagement Rate (%)']

print("Average Persen Engagement Rate of Music Content:",Music_engagement.mean())
print("Average Persen Engagement Rate of Photograph Content:",Photography_engagement.mean())
print("Average Persen Engagement Rate of Cinema Content:",Cinema_engagement.mean())
```

```python
f_stat,p_value = stats.f_oneway(Music_engagement, Photography_engagement, Cinema_engagement)
print('P-value:',p_value)
```

Berdasarkan uji tersebut didapat nilai p-value > 0.05, artinya H0 pada uji hipotesa ini gagal untuk ditolak. hal ini berarti tidak ada perbedaan yang signifikan antara ketiga jenis konten tersebut dengan engagement rate. dengan kata lain, jenis konten tidak berpengaruh terhadap engagement rate suatu konten atau apapun kontennya akan menghasilkan engagement rate yang sama aja dari ketiga jenis konten tersebut.

jadi buat kalian yang ingin jadi konten kreator baru bebas memilih konten dari ketiga jenis konten tersebut untuk memulai karir sebagai seorang influencer

- **Menentukan Hubungan antara Jumlah Followers dengan Engagement Rate dengan mengunakan uji T-Test dua sampel**

    H0 : μ jumlah followers  =  μ engagement rate
    H1 : μ jumlah followers !=  μ engagement rate

```python
avg_followers = dataig_new['Followers']
avg_engagement = dataig_new['Persen Engagement Rate (%)']

print('Average followers di Instagram:', avg_followers.mean())
print('Average engagement di Instagram:', avg_engagement.mean())
```

```python
t_stat, p_val = stats.ttest_ind(avg_followers, avg_engagement)
print('P-value:',p_val) 
print('t-statistics:',t_stat)
```

Berdasarkan hasil uji tersebut didapat bahwa nilai p-value kurang dari 0.05 yang artinya H0 pada uji hipotesa ini ditolak. hal ini berarti tidak ada hubungan antara pertambahan jumlah followers dengan engagement rate. dengan kata lain, orang yang melihat postingan di ig belum tentu men-follow konten kreator yang membuat postingan tersebut sedangkan ketika audience tersebut melihat postingan tersebut telah menambah persen engagement rate pada konten kreator tersebut.
